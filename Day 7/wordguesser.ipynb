```python
import random
from IPython.display import Image, display

wordList = ["rizz", "ohio", "sigma", "tiktok", "skibidi"]


attempts = 10



while attempts > 0:
    word = random.choice(wordList)
    guessedWord = "_" * len(word)
    print("\nCurrent word: " + ' '.join(guessedWord))
    user_guess = str(input("Whats your guess? ").lower())
    if word == user_guess:
        print("You win! Congrats!")
        display(Image(filename="congrats.jpg"))
        break
    else:
        attempts -= 1
        print(f"Your Remaining Attempt: {attempts}")
        print("Wrong Answer! New Guess word: ")
        continue
```
