```python
import os
import random
from typing import TypedDict, List
from dotenv import load_dotenv
from IPython.display import Image, display
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI
from langgraph.graph import StateGraph, START, END
```

```python
load_dotenv()
apikey = os.getenv("OPENAI_API_KEY")

if not apikey:
    raise ValueError("❌ Missing OPENAI_API_KEY in your .env file.")


# ✅ Initialize the LLM (for OpenRouter)
llm = ChatOpenAI(
    model="gpt-4o-mini",
    api_key=os.environ["OPENAI_API_KEY"],
    base_url="https://openrouter.ai/api/v1"
)
```

```python
class AgentState(TypedDict):
    message: List[HumanMessage]


# ✅ Define the function node
def rate(state: AgentState) -> AgentState:
    systemprompt = SystemMessage(
        content=(
            "You are a rizz calculator expert. Rate the user's rizz from 1–10, "
            "and include a short roast or compliment using Gen Z and brainrot slang. "
            "Keep it short — 1–2 sentences only."
        )
    )

    all_messages = [systemprompt] + state["message"]
    response = llm.invoke(all_messages)

    print(f"\n🧠 Rizz Bot: {response.content}\n")

    memes = [
        'mem1.jpg', 'mem2.jpg', 'mem3.jpg', 'mem4.jpg',
        'mem5.jpg', 'mem6.jpg', 'mem7.jpg', 'mem8.jpg', 'mem9.jpg'
    ]
    random_memes = random.choice(memes)
    try:
        display(Image(filename=random_memes))
    except Exception as e:
        print(f"(⚠️ Couldn't display meme: {e})")

    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("start_rizz", rate)
graph.add_edge(START, "start_rizz")
graph.add_edge("start_rizz", END)

bot = graph.compile()


# ✅ Interactive loop
while True:
    user_input = input("Enter a Message to be Roasted (or type 'exit' to quit): ")
    if user_input.lower() == "exit":
        print("👋 Bye! Stay rizzful.")
        break

    result = bot.invoke({"message": [HumanMessage(content=user_input)]})
```
